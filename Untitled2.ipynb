```python
!pip install python-telegram-bot==13.15

import logging
import random
from telegram import Update
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters, CallbackContext

# üîê W≈Çasny token z @ApteczkaDDLBOT
TOKEN = "8182943260:AAFYRmPOFHZxUlSVciP5FAFJ9g9UpMXEdo"
 # ‚Üê Zmie≈Ñ na sw√≥j

# Usuniƒôto ≈∫le wciƒôte i nieprzypisane cytaty

# Logi
logging.basicConfig(
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    level=logging.INFO
)

# Komenda /start
def start(update: Update, context: CallbackContext) -> None:
    update.message.reply_text("Hej... czemu mnie budzisz? üò©")

# Odpowied≈∫ losowa w stylu Piwnicy
def echo(update: Update, context: CallbackContext) -> None:
    odpowiedz = random.choice(ODPOWIEDZI)
    update.message.reply_text(odpowiedz)

# G≈Ç√≥wna funkcja
def main():
    # Stworzenie instancji Updatera i Dispatchera
    updater = Updater(TOKEN, use_context=True)
    dp = updater.dispatcher

    # Komendy Piwnicy
    dp.add_handler(CommandHandler("start", start))
    # Upewnij siƒô, ≈ºe funkcje status, skrot, aparta, fredro, krzysiek, drop, adam, kebabdom, piwnica sƒÖ zdefiniowane
    # W obecnym kodzie nie sƒÖ, co spowoduje NameError. Doda≈Çem przyk≈Çadowe definicje poni≈ºej.
    # Je≈õli masz ju≈º te funkcje w innych kom√≥rkach, mo≈ºesz usunƒÖƒá te przyk≈Çadowe.
    # Pamiƒôtaj, aby wszystkie funkcje handler√≥w by≈Çy zdefiniowane PRZED wywo≈Çaniem main().
    # dp.add_handler(CommandHandler("status", status)) # Funkcja 'status' nie zdefiniowana
    #dp.add_handler(CommandHandler("skrƒôt",  skrƒôt))
    dp.add_handler(CommandHandler("aparta", aparta))
    dp.add_handler(CommandHandler("fredro", fredro))
    dp.add_handler(CommandHandler("krzysiek", krzysiek))
    dp.add_handler(CommandHandler("drop", drop))
    dp.add_handler(CommandHandler("adam", adam))
    dp.add_handler(CommandHandler("kebabdom", kebabdom))
    dp.add_handler(CommandHandler("piwnica", piwnica))

    # Odpowied≈∫ na zwyk≈Çe wiadomo≈õci (opcjonalnie)
    dp.add_handler(MessageHandler(Filters.text & ~Filters.command, main_handler))
    # Zauwa≈ºy≈Çem, ≈ºe sekcja main jest powt√≥rzona. UsunƒÖ≈Çem duplikat i po≈ÇƒÖczy≈Çem handlery.

    #updater.start_polling()
    print("üü¢ Bot dzia≈Ça i m√≥wi jak z Piwnicy!")
    updater.idle()

# Przyk≈Çadowe definicje funkcji handler√≥w, je≈õli nie sƒÖ zdefiniowane gdzie indziej
# Je≈õli masz je zdefiniowane w innych kom√≥rkach, usu≈Ñ poni≈ºszy blok
def status(update: Update, context: CallbackContext):
    update.message.reply_text("Status: Dzia≈Çam!")

def skrot(update: Update, context: CallbackContext):
    update.message.reply_text(random.choice(SKRƒòT_QUOTES))

def aparta(update: Update, context: CallbackContext):
    update.message.reply_text(random.choice(APARTA_QUOTES))

def fredro(update: Update, context: CallbackContext):
    update.message.reply_text(random.choice(FREDRO_QUOTES))

def krzysiek(update: Update, context: CallbackContext):
    update.message.reply_text(random.choice(KRZYSIEK_QUOTES))

def drop(update: Update, context: CallbackContext):
    update.message.reply_text(random.choice(DROP_QUOTES))

def adam(update: Update, context: CallbackContext):
    update.message.reply_text(random.choice(ADAM_QUOTES))

def kebabdom(update: Update, context: CallbackContext):
    update.message.reply_text(random.choice(KEBABDOM_QUOTES))

def piwnica(update: Update, context: CallbackContext):
    update.message.reply_text(random.choice(ODPOWIEDZI))

def main_handler(update: Update, context: CallbackContext):
    text = update.message.text.lower()
    if any(s≈Çowo in text for s≈Çowo in ["g≈Çodny", "zmƒôczony", "nie mam", "spalony"]):
        update.message.reply_text(random.choice(KEBABDOM_QUOTES))
    else:
        update.message.reply_text(random.choice(ODPOWIEDZI))  # fallback


# --- Historie Piwnicy ---
# Upewnij siƒô, ≈ºe te listy sƒÖ zdefiniowane PRZED wywo≈Çaniem funkcji, kt√≥re ich u≈ºywajƒÖ (np. main)
SKRƒòT_QUOTES = [
    "Mam skrƒôta czy nie mam ju≈º?",
    "To by≈Ç ostatni skrƒôt... ≈ºycie stracone.",
    "Nie wiem, co jeszczesz gadaƒá... wszystko wiruje."
]
APARTA_QUOTES = [
    "Nie ma nic na Aparta, wyrzucajƒÖ nas!",
    "Kto zamknƒÖ≈Ç drzwi?! Adam! Otw√≥rz!",
    "W Apartach jest gorzej ni≈º w mojej g≈Çowie."
]

FREDRO_QUOTES = [
    "Oddajesz mi apteczkƒô. Teraz.",
    "Fredro nie pyta. Fredro patrzy. Fredro zabiera apteczkƒô.",
    "Za s≈Çaby temat. Apteczka wraca do mnie."
]

KRZYSIEK_QUOTES = [
    "Da≈Çem iPad za 70 z≈Ç... bo nie by≈Ço co je≈õƒá. Teraz chcƒô 120 z≈Ç.",
    "iPad to nie tylko gad≈ºet. To symbol naszej wsp√≥lnej traumy.",
    "Nie jestem wkurzony. Jestem strategiczny."
]

DROP_QUOTES = [
    "Adam jedzie po dropa!",
    "Drop by≈Ç... ale nie ten.",
    "Nie ma ju≈º ≈ºadnego dropa. Tylko puste opakowania."
]

ADAM_QUOTES = [
    "Jadƒô po dropa... ale nie wiem gdzie to jest.",
    "Makatusin dzia≈Ça na wszystko. Nawet na dryf.",
    "Fredro mnie ignoruje. Krzysiek chce iPad. A ja chcƒô tylko... normalno≈õci."
]

KEBABDOM_QUOTES = [
    "Mieszkam w Kebab Dom. To nie jest kebab. To jest katedra.",
    "Nie pamiƒôtam, jak tu trafi≈Çem. Ale bu≈Çka by≈Ça ciep≈Ça.",
    "W Kebab Dom nie ma regu≈Ç. Tylko menu. I jeszcze jeden st√≥≈Ç, kt√≥ry siƒô chwieje."
]

# Ca≈Ça lista do losowych odpowiedzi
ODPOWIEDZI = SKRƒòT_QUOTES + APARTA_QUOTES + FREDRO_QUOTES + KRZYSIEK_QUOTES + DROP_QUOTES + ADAM_QUOTES + KEBABDOM_QUOTES


if __name__ == '__main__':
    main()
```
